Escreva uma função chamada **compara** que receba como parâmetro uma pilha e uma fila. Ela deve retornar True caso as duas estruturas passadas como parâmetro sejam iguais e False caso contrário. Um pilha e uma fila são consideradas iguais tanto quando os elementos estão na mesma ordem ou em ordem contrária. Por exemplo, a pilha A=[0,1,2,3,4] é igual as filas B = [0,1,2,3,4] e C = [4,3,2,1,0].

 - Você deve usar as classes Pilha e Fila ensinadas na disciplina e disponibilizadas juntamente com este notebook.

- As únicas operações permitidas para acessar os valores da pilha e fila são, respectivamente, as operações pop e dequeue.

- Você deve garantir que ao final da função, a pilha e fila passadas como parâmetro estejam intactas, isto é, como os mesmos valores passados como entrada.

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
from utils.pilhaDinamica import *
from utils.filaDinamica import *

# Definindo a função 'compara' com dois parâmetros: uma pilha e uma fila
def compara(pilha, fila):

  # Criando uma nova pilha para armazenar os elementos da fila
  pilhaAuxFila = Pilha()

  # Inicializando as variáveis que vão armazenar os elementos das pilha e da fila
  numPilha = 0
  numFila = 0

  # Inicializando a variável 'saoIguais' como falsa
  saoIguais = False

  # Criando duas novas estruturas de dados para armazenar as pilhas e filas originais
  pilhaSalva = Pilha()
  filaSalva = Fila()

  # Criando uma nova pilha para armazenar os elementos da pilha invertida
  pilha_inversa = Pilha()

  # Criando uma nova fila para armazenar os elementos da fila original
  fila2 = Fila()

  # Verificando se o tamanho da pilha é diferente do tamanho da fila
  if pilha.size() != fila.getSize():
      return False

  # Loop para verificar se os elementos da pilha e da fila são iguais
  while not pilha.isEmpty():

    # Desempilhando o último elemento da pilha e removendo o primeiro elemento da fila
    numPilha = pilha.pop()
    numFila = fila.dequeue()

    # Verificando se os elementos são iguais
    if numPilha == numFila:
      saoIguais = True
    else:
      saoIguais = False

    # Salvando os elementos da pilha e da fila em suas respectivas estruturas de dados
    pilhaSalva.push(numPilha)
    filaSalva.enqueue(numFila)

    # Salvando os elementos da pilha invertida e da fila original em suas respectivas estruturas de dados
    pilha_inversa.push(numPilha)
    fila2.enqueue(numFila)

  # Verificando se os elementos da pilha e da fila são iguais
  if saoIguais:
    # Loop para empilhar os elementos da pilhaSalva e enfileirar os elementos da filaSalva na pilha e na fila original
    while not pilhaSalva.isEmpty():
      pilha.push(pilhaSalva.pop())
      fila.enqueue(filaSalva.dequeue())
    return True
  else:
    # Loop para verificar se os elementos da pilha invertida e da fila2 são iguais
    while not pilha_inversa.isEmpty():
      if pilha_inversa.pop() == fila2.dequeue():
        saoIguais = True
      else:
        saoIguais = False

      # Empilhando os elementos da pilhaSalva e enfileirando os elementos da filaSalva na pilha e na fila original
      pilha.push(pilhaSalva.pop())
      fila.enqueue(filaSalva.dequeue())

    # Verificando se os elementos da pilha invertida e da fila2 são iguais
    if saoIguais:
      return True
    else:
      return False

##########################################################################

# testando a função criada
pilha1 = Pilha()
pilha1.push(0)
pilha1.push(1)
pilha1.push(2)
pilha1.push(3)
pilha1.push(4)

# testando a função criada
pilha2 = Pilha()
pilha2.push(4)
pilha2.push(3)
pilha2.push(2)
pilha2.push(1)
pilha2.push(0)

fila1 = Fila()
fila1.enqueue(0)
fila1.enqueue(1)
fila1.enqueue(2)
fila1.enqueue(3)
fila1.enqueue(4)

fila2 = Fila()
fila2.enqueue(4)
fila2.enqueue(3)
fila2.enqueue(2)
fila2.enqueue(1)
fila2.enqueue(0)

fila3 = Fila()
fila3.enqueue(4)
fila3.enqueue(3)
fila3.enqueue(2)
fila3.enqueue(0)
fila3.enqueue(1)

comp1 = compara(pilha1, fila1)
comp2 = compara(pilha1, fila2)
comp3 = compara(pilha2, fila1)
comp4 = compara(pilha2, fila2)
comp5 = compara(pilha2, fila3)

print('Resultado retornado:')
print('\tComparacao 1?: ', comp1)
print('\tComparacao 2?: ', comp2)
print('\tComparacao 3?: ', comp3)
print('\tComparacao 4?: ', comp4)
print('\tComparacao 5?: ', comp5)

print('\n' + 20*'-')
print('Resultado esperado:')
print('\tComparacao 1?: True')
print('\tComparacao 2?: True')
print('\tComparacao 3?: True')
print('\tComparacao 4?: True')
print('\tComparacao 5?: False')
````

## Exercício 2

Faça uma classe chamada **ListaNSE**. Essa classe deve implementar uma **lista não sequencial e estática**. Essa lista deve ter o mesmo comportamento de uma **lista não sequencial e dinâmica**, mas deve ser guardada dentro de uma lista em Python com capacidade limitada.

Essa classe deve possui os seguintes métodos:

- inserir(x, i): insere x na posição i da estrutura
- remover(i): remove o item que está na posição i da estrutura e retorna seu valor
- buscar(x): buscar o item x na lista e retorna sua posição
- size(): retorna o número de elementos da lista
- clear(): remove os elementos da lista
- \_\_str\_\_(): imprime a estrutura

- Você deve usar a lista do Python apenas para inicializar a estrutura. Depois disso, nenhum método do Python pode ser usado, como append, insert, etc.

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

class Node:
  """
  Nó de uma lista duplamente ligada
  """
  def __init__(self, data = None, next = None):
      self.data = data
      self.next = next # apontador para o nó da frente

class ListaNSE:
  def __init__(self, capacidade):
      self.capacidade = capacidade
      self.head = None
      self.dados = [None] * capacidade  # cria uma lista de None com o tamanho da capacidade

  def insert(self, valor, posicao):
    if posicao > self.capacidade:
        print('Posição inválida')  # levanta um erro se a posição for maior que a capacidade

    novo_no = Node(valor)

    if self.head is None:
        # Se a lista está vazia, o novo nó se torna a cabeça da lista
        self.head = novo_no
    elif posicao == 0:
        # Se a posição é 0, o novo nó é inserido no início da lista
        novo_no.next = self.head
        self.head = novo_no
    else:
        # Percorre a lista até encontrar a posição correta
        temp = self.head
        for i in range(posicao - 1):
            if temp is None:
                # Se já percorreu toda a lista e ainda não chegou na posição,
                # levanta um erro
                print('Posição inválida')
            temp = temp.next

        # Insere o novo nó na posição correta
        novo_no.next = temp.next
        temp.next = novo_no

    # Insere o valor na lista de dados
    self.dados[posicao] = valor  # adiciona o valor na posição da lista de dados correspondente

  def remove(self, i):
    for j in range(len(self.dados)):
      if self.dados[j] == i:
        self.dados[j] = None  # remove o valor correspondente da lista de dados

  def busca(self, item):
    count = 0
    for i in self.dados:
      count+=1
      if i == item:
        return count  # retorna a posição do item se ele for encontrado

  def size(self):
    count = 0
    for i in self.dados:
      if i is not None:
        count += 1
    return count  # conta quantos elementos estão na lista

  def clear(self):
    self.head = None
    self.size = 0
    self.dados = [None] * self.capacidade  # reseta a lista para um estado vazio

  def __str__(self):
    """
    Retorna uma string com as informações da lista
    quando o objeto é chamado dentro de um print()
    ou dentro de um str().

    A lista não possui restrição de acesso aos seus elementos.
    Por isso, ela não precisa ser destruída para que seus elementos sejam
    visitados como na fila e na pilha.
    """
    if self.head is None:
        return '[ ]'

    strfila = '['
    # começando pelo head
    tempNode  = self.head
    valor = tempNode.data

    # guarda o valor do head em uma string
    strfila += str(valor)

    # percorre o resto da lista até que o apontador next do nó seja nulo
    # o que indica que chegou no tail
    while tempNode.next is not None:

        # vai para o proximo noh
        tempNode = tempNode.next

        valor = tempNode.data

        # guarda o valor em uma string
        strfila += ',' + str(valor)

    strfila += ']'

    return strfila

##########################################################################

# testa a Lista
lista = ListaNSE(5)

# inserir no inicio da lista
lista.insert(5,0)
print( 'insert(5,0): ', lista )

lista.insert(10,0)
print( 'insert(10,0): ', lista )

lista.insert(24,2)
print( 'insert(24,2): ', lista )

lista.insert(25,1)
print( 'insert(25,1): ', lista )

lista.remove(2)
print( 'remove(2): ', lista )

print( 'Size: ', lista.size() )

lista.clear()
print( 'Clear: ', lista )

print('\n' + 20*'-')
print('Resultado esperado:')
print( 'insertHead(5,0): [5]')
print( 'insertHead(10,0): [10,5]')
print( 'insert(24,2): [10,5,24]')
print( 'insert(25,1): [10,25,5,24]')
print( 'remove(2): [10,25,24]')
print( 'Size: 3')
print( 'Clear: []')
````

O critério para você ordenar as strings dessa lista é um **número inteiro** que represente a quantidade média de caracteres de cada palavra arredondada para baixo.
````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

class No:
  def __init__(self, data=None):
    self.data = data
    self.mediaC = None
    self.next = None
    self.prev = None

class ListaInsertion:
  def __init__(self):
    self.head = None
    self.tail = None
    self.size = 0

  def __str__(self):
    strfila = '['

    if self.size > 0:
      # comecao pelo head
      tempNode  = self.head
      data = tempNode.data

      # guarda o data do head em uma string
      strfila += str(data) + ","

      # percorre o resto da lista ate que o apontador next do noh seja nulo
      # o que indica que chegou no tail
      while tempNode.next is not None:

          # vai para o next noh
          tempNode = tempNode.next

          data = tempNode.data

          # guarda o data em uma string
          if tempNode.next == None:
              strfila += ' ' + str(data)
          else:
              strfila += ' ' + str(data) + ","

    strfila += ']'

    return strfila

  def insert(self, data):
    # Cria um novo nó com o dado recebido como argumento
    novo_no = No(data)

    # Calcula a média do tamanho das palavras do dado
    count = 0
    size_medio1 = 0
    for palavra in data.split():
        count += 1
        size_medio1 += len(palavra)
    novo_no.mediaC = int(size_medio1/count)

    # Se a lista estiver vazia, o novo nó é o head e o tail
    if self.size == 0:
        self.head = novo_no
        self.tail = novo_no
    else:
        no_atual = self.head
        while no_atual and self.comparar(data, no_atual.data) == True:
            # Percorre a lista até encontrar a posição correta para inserir o novo nó
            no_atual = no_atual.next

        if no_atual:
            if no_atual == self.head:
                if no_atual.mediaC == novo_no.mediaC:
                    # Insere o novo nó imediatamente após o nó com a mesma média de tamanho
                    novo_no.next = self.head.next
                    novo_no.prev = self.head
                    self.head.next.prev = novo_no
                    self.head.next = novo_no
                else:
                    # Insere o novo nó no início da lista
                    novo_no.next = self.head
                    self.head.prev = novo_no
                    self.head = novo_no
            else:
                # Insere o novo nó no meio da lista
                novo_no.next = no_atual
                novo_no.prev = no_atual.prev
                no_atual.prev.next = novo_no
                no_atual.prev = novo_no
        else:
            # Insere o novo nó no final da lista
            novo_no.prev = self.tail
            self.tail.next = novo_no
            self.tail = novo_no

    # Incrementa o tamanho da lista
    self.size += 1

  def comparar(self, string1, string2):
    # Calcula a média do tamanho das palavras da primeira string
    count = 0
    size_medio1 = 0
    for palavra in string1.split():
        count += 1
        size_medio1 += len(palavra)
    size_medio1 = int(size_medio1/count)

    # Calcula a média do tamanho das palavras da segunda string
    count = 0
    size_medio2 = 0
    for palavra in string2.split():
        count += 1
        size_medio2 += len(palavra)
    size_medio2 = int(size_medio2/count)

    # Retorna True se as duas strings possuem a mesma média de tamanho das palavras
    return size_medio1 == size_medio2

  def repeticoes(self, numero):

    return self.countOccurrences(numero)

  def binarySearch(self, l, r, x):
    # Se o índice direito é menor que o índice esquerdo, o elemento não foi encontrado
    if (r < l):
        return -1

    # Calcula o índice do meio da lista
    mid = int(l + (r - l) / 2)

    # Encontra o nó que corresponde ao índice do meio
    meio = self.head
    c = 0
    while c != mid:
        meio = meio.next
        c += 1

    # Verifica se o valor do nó do meio é igual ao valor procurado
    if meio.mediaC == x:
        return mid

    # Se o valor do nó do meio é maior que o valor procurado,
    # faz uma busca binária na metade esquerda da lista
    if meio.mediaC > x:
        return self.binarySearch(l, mid - 1, x)

    # Se o valor do nó do meio é menor que o valor procurado,
    # faz uma busca binária na metade direita da lista
    return self.binarySearch(mid + 1, r, x)

  def countOccurrences(self, x):
    # Encontra o índice do primeiro nó com médiaC igual a x usando busca binária
    ind = self.binarySearch(0, self.size - 1, x)

    # Se o elemento não foi encontrado, retorna 0
    if ind == -1:
        return 0

    # Inicializa um contador de ocorrências e procura para a esquerda e para a direita do nó encontrado
    count = 0
    left = ind - 1
    if left >= 0:
        esquerda = self.head
        c = 0
        while c != left:
            esquerda = esquerda.next
            c += 1
    while (left >= 0 and esquerda.mediaC == x):
        count += 1
        left -= 1
        esquerda = esquerda.prev

    # Incrementa o contador de ocorrências para o nó encontrado
    count += 1

    right = ind + 1
    if right < self.size:
        direita = self.head
        d = 0
        while d != right:
            direita = direita.next
            d += 1
    while (right < self.size and direita.mediaC == x):
        count += 1
        right += 1
        direita = direita.next

    # Retorna o número de ocorrências encontradas
    return count

  # Define o método sublista() na classe ListaInsertion
  def sublista(self, a, b):
    # Define o nó atual como a cabeça da lista
    current_node = self.head
    # Percorre a lista até o índice "a"
    for i in range(a):
        # Se o nó atual for None (ou seja, não há mais elementos na lista), retorna None
        if current_node is None:
            return None
        # Define o nó atual como o próximo nó da lista
        current_node = current_node.next
    # Cria uma nova lista vazia
    new_list = ListaInsertion()
    # Define a cabeça da nova lista como o nó atual
    new_list.head = current_node
    # Percorre a lista do índice "a" até o índice "b"
    for i in range(a, b):
        # Se o nó atual for None, retorna a nova lista
        if current_node is None:
            return new_list
        # Define o nó atual como o próximo nó da lista
        current_node = current_node.next
        # Insere o valor do nó atual na nova lista
        new_list.insert(current_node.data)
    # Retorna a nova lista
    return new_list



##########################################################################

# testa o exercicio
lista = ListaInsertion()

# insert no inicio da lista
lista.insert("Centro Universitario Facens")
lista.insert("Eu fui para um shopping em Sorocaba")
lista.insert("A menina gosta de jogar")
lista.insert("O menino gosta de andar")
lista.insert("Tá tá tá")

print("\nLista:\n", lista)

qtd1 = lista.repeticoes(3)
qtd2 = lista.repeticoes(2)
qtd3 = lista.repeticoes(30)
print("Qtd. repeticoes 1: ", qtd1)
print("Qtd. repeticoes 2: ", qtd2)
print("Qtd. repeticoes 3: ", qtd3)

print('\n' + 20*'-')
print('Resultado esperado:')
print("\nLista:\n")
print('[Tá tá tá, A menina gosta de jogar, ', end="")
print('O menino gosta de andar, ', end="")
print('Eu fui para um shopping em Sorocaba, ', end="")
print('Centro Universitario Facens]', end="")
````
